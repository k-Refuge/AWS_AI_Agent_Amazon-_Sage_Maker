```python
# test_endpoint.py
import boto3
import json
from botocore.config import Config

# -------------------------------
# ⚙️ Configurer le client SageMaker Runtime avec timeout
# -------------------------------
my_config = Config(
    read_timeout=300,      # 5 minutes pour la lecture de réponse
    connect_timeout=60     # 1 minute pour établir la connexion
)

runtime = boto3.client(
    "sagemaker-runtime",
    region_name="us-east-1",
    config=my_config
)

# -------------------------------
# 🔹 Nom de ton endpoint
# -------------------------------
endpoint_name = "legal-agent-endpoint-9"

# -------------------------------
# 🔹 Payload d'exemple à envoyer
# -------------------------------
payload = {
    "question": "Bonjour, je veux contester mon licenciement."
}

# -------------------------------
# 🔹 Appel à l'endpoint
# -------------------------------
try:
    response = runtime.invoke_endpoint(
        EndpointName=endpoint_name,
        ContentType="application/json",
        Body=json.dumps(payload)
    )

    result = response["Body"].read().decode("utf-8")
    print("✅ Réponse du modèle :")
    print(result)

except Exception as e:
    print("❌ Erreur lors de l'appel à l'endpoint :")
    print(str(e))
```
